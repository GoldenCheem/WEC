```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('World Energy Consumption.csv')
df
```

```python
col = ['iso_code', 'country', 'year', 'gdp', 'population', 'fossil_fuel_consumption', 'coal_consumption', 'coal_production', \
       'gas_consumption', 'gas_production', 'oil_consumption', 'oil_production', 'low_carbon_consumption', 'renewables_consumption', \
       'biofuel_consumption', 'hydro_consumption', 'solar_consumption', 'wind_consumption', 'other_renewable_consumption', \
       'nuclear_consumption']

df = df[col]
df
```

```python
df = df.dropna(subset=['iso_code'], how='any')
df
```

```python
df = df[(df['year'] >= 2000) & (df['year'] <= 2016)].reset_index().iloc[:,1:]
df = df.fillna(0)
df.head(50)
```

```python
df = df[df['gdp'] != 0]
df
```

```python
df = df[(df.fossil_fuel_consumption != 0) | (df.low_carbon_consumption != 0)].reset_index().iloc[:,1:]
df.to_csv('energy_consumption.csv', index=False)
df
```
