```python
df5 = df[['country', 'year', 'gdp', 'population', 'fossil_fuel_consumption', 'low_carbon_consumption']].iloc[:1343, :]
df5
```

```python
X = df5[df5.year == 2016][['country', 'gdp']]
X['gdp'] = X['gdp'].apply(lambda x: np.log10(x))
X
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X[['gdp']])
labels = kmeans.labels_
# [print(labels[i]) for i in range(len(labels))]
np.unique(labels, return_counts=True)
```

```python
train_data = df5[df5.year <= 2015]
test_data = df5[df5.year == 2016]

X_train = train_data[['population' , 'fossil_fuel_consumption', 'low_carbon_consumption']].values
y_train = train_data['gdp'].values

X_test = test_data[['population' , 'fossil_fuel_consumption', 'low_carbon_consumption']].values
y_test = test_data['gdp'].values

model = LinearRegression().fit(X_train, y_train)
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

```python
result = pd.DataFrame(list(zip(test_data['country'].values, y_pred, y_test, labels)),
                      columns=['country', 'predicted', 'actual', 'group'])
result['group'] = result['group'].apply(lambda x: 'High GDP' if x == 0 else 'Medium GDP' if x == 1 else 'Low GDP')
result.to_csv('linear_regression.csv', index=False)
result
```
